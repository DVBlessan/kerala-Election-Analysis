<h1 style="font-family:newtimeroman;font-size:130%;text-align:left">District wise Total Votes Polled </h1>

```python
total=pd.DataFrame(df.groupby('DISTRICT')['TOTAL VOTES'].sum())
total.plot.bar(color='orange')
plt.xlabel("DISTRICT",size=15)
plt.ylabel("TOTAL VOTES",size=15)
plt.title("Kerala District wise Total Votes Polled",size=20)
legend=plt.legend()
legend.remove()
```

```python
plt.figure(figsize=(9,10))
ldf=df['LDF'].sum()
udf=df['UDF'].sum()
nda=df['NDA'].sum()
oth=df['OTHERS'].sum()
nota=df['NOTA'].sum()
x=[ldf,udf,nda,oth,nota]
plt.pie(x,labels=['LDF','UDF','NDA','OTHERS','NOTA'],autopct='%1.1f%%',textprops={'fontsize': 14},shadow=True)
plt.title("Total votes for each parties and nota in Kerala Assembly Election",size=20)
plt.legend()
plt.show()
```

```python
evm=df['EVM VOTES'].sum()
post=df['POSTAL VOTES'].sum()
y=[evm,post]
plt.pie(y,labels=['EVM VOTES','POSTAL VOTES'],autopct='%1.1f%%',textprops={'fontsize': 14},shadow=True)
plt.title("Total EVM Votes & Postal Votes",size=20)
plt.legend()
plt.show()
```

```python
totaldf=pd.DataFrame(df.groupby('DISTRICT')['LDF','UDF','NDA','NOTA','OTHERS'].sum())
totaldf.plot.bar(width=0.8)
plt.title("District wise Parties Votes",size=20)
plt.ylabel("Votes")
plt.show()
```

```python
df.DISTRICT.value_counts().plot.pie(autopct='%1.1f%%',textprops={'fontsize': 14},shadow=True)
plt.legend()
plt.title("Constituency counts in each District in kerala for Assembly Election",size=25)
plt.show()
```

```python
votedf=pd.DataFrame(df.groupby('DISTRICT')['EVM VOTES','POSTAL VOTES'].sum())
votedf.plot.bar(width=0.8)
plt.title("District wise EVM & Postal Votes in Kerala Assembly Election",size=20)
plt.ylabel("Votes")
plt.show()
```

```python
df.WINNER.value_counts().plot.pie(autopct='%1.1f%%',textprops={'fontsize': 14},shadow=True)
plt.title("Winner for the Kerala Assembly Election 2021",size=25)
plt.legend
plt.show()
```

```python
cordf=pd.DataFrame()
cordf=df.drop(['CONSTITUENCY','DISTRICT','WINNER','LDF','UDF','NDA','NOTA','OTHERS','EVM VOTES','POSTAL VOTES','TOTAL VOTES','LEAD'],axis=1)
plt.title("Heatmap for the % of Kerala election",size=20)
cordf=cordf.reset_index(drop=True)
cor=cordf.corr()
sbn.heatmap(cor,annot=True)
plt.show()
```

```python
sbn.jointplot(df['TOTAL VOTES'],df['LEAD'])
plt.show()
```
